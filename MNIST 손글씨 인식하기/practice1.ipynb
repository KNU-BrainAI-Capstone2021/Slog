```python
from keras.datasets import mnist#MNISt데이터는 케라스를 이용해 불러옴.
from keras.utils import np_utils

import numpy
import sys
import tensorflow as tf

#seed 값을 설정하겠다.
seed=0
numpy.random.seed(seed)
tf.random.set_seed(3)

#MNIST 데이터셋을 불러오겠다.
(X_train, Y_class_train), (X_test, Y_class_test) = mnist.load_data()

print("학습셋 이미지 수 : %d 개" %(X_train.shape[0]))
print("테스트셋 이미지 수 : %d 개" %(X_test.shape[0]))
```

```python
#불러온 이미지 중에서 한개의 이미지를 그래프로 확인하려고 하는데 첫번째 데이터를 출력하겠다.
import matplotlib.pyplot as plt
plt.imshow(X_train[0], cmap='Greys')
plt.show()
```

```python
#코드로 확인하겠다.
for x in X_train[0]:
    for i in x:
        sys.stdout.write('%d\t' %i)
    sys.stdout.write('\n')
```

```python
#가로 28, 세로 28의 2차원 배열을 784개의 1차원 배열로 바꿔 주겠다.
X_train = X_train.reshape(X_train.shape[0], 784)
X_train = X_train.astype('float64')
X_train = X_train/255
#X_test에도 적용하겠다. 단, 한꺼번에 적용하는 방법으로 하겠다.
X_test = X_test.reshape(X_test.shape[0], 784).astype('float64')/255

#클래스 값을 확인하여 숫자 이미지에 매겨진 이름을 확인하겠다.
print("class : %d "%(Y_class_train[0]))
```
